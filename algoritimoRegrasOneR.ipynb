# Matriz de erros


```python
def calculaMelhores(dados,X,Y,indice):
  temp=[]
  for x in X:
    for y in Y:
      temp.append((x,y,dados[(dados.iloc[:,indice]==x) & (dados.iloc[:,-1]==y) ].shape[0]))

  melhores=pd.DataFrame(temp)
  print(melhores)

  maiores_indices=[]
  for x in X:
    temp2=melhores[melhores.iloc[:,0]==x]
    col=temp2.iloc[:,-1]
    maiores_indices.append(col.idxmax())
  
  return melhores.iloc[maiores_indices,:]

def GeraAnaliseDeErros(dados,colunas,indice,classe):#gera um dataframe por par variavel classe e depois Juntamos os dataframes em um unico
  var1 = colunas[indice]
  X=dados.loc[:,[var1]]
  Y=classe
  ####Pegando atributo CHECK

  ####Pegando as quantidades CHECK
  rel_tipos_quantidadesX=np.unique(X,return_counts=True)[0]
  rel_tipos_quantidadesY=np.unique(Y,return_counts=True)[0]
 ##Pegando o numero de quantidades CHECK
  rel_numero_quantidadesX=np.unique(X,return_counts=True)[1]

  #####Efetuando o calculo com as possibilidades CHECK
  dataMelhores=calculaMelhores(dados,rel_tipos_quantidadesX,rel_tipos_quantidadesY,indice)

  #####Efetuando o calculo do ERRO 
  #(quantidade- melhor valor escolhido)/quantidade 
  MVES=[]
  sum_valor_menor_melhor=0
  zipados_qtd_tipo=zip(rel_tipos_quantidadesX,rel_numero_quantidadesX)
  for tipo , valor in zipados_qtd_tipo:
    melhor_valor_escolhido=dataMelhores[dataMelhores.iloc[:,0]==tipo].iloc[:,-1]
    #print(melhor_valor_escolhido)
    termo_subtrator=int(valor-melhor_valor_escolhido)
    
    sum_valor_menor_melhor=sum_valor_menor_melhor+termo_subtrator
    MVES.append(termo_subtrator/valor)
   ######ERROS TOTAIS
  erro_total=sum_valor_menor_melhor/Y.shape[0]

  return dataMelhores,erro_total
  #matrix=pd.DataFrame({'Atributo':var1,'Quantidade':[],'N_Quantidade':[],'Regras':[],'Erro':[],'Total_Erro':[]})

```

```python
for i in range(len(colunas)-1):
  print(GeraAnaliseDeErros(dados,colunas,i,dados.iloc[:,-1]))
```
